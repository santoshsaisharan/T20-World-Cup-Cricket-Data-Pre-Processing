<h1 style="color:green">Converting the json files into csv files & preprocessing the data</h1>

<h2 style="color:red">Match results data</h2>

```python
data = pd.read_json('t20_json_files/t20_wc_match_results.json')
#the data df contains and additional subset of json file so we have to get the inner json file.
data['matchSummary'][0]
```

```python
df_match = pd.DataFrame(data['matchSummary'][0])
df_match.head()
#Now we have all the match data in the df_match dataframe
```

```python
df_match.shape
```

```python
len(df_match['scorecard'].unique())
# here we can see that we have 45 rows in this df and there 45 unique scorecard values. Hence, I wanna treat
# the scorecard column as the unique id. 
```

```python
# Lets rename the scorecard to match_id
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match.to_csv('t20_json_files/match_results_transformed.csv', index= False)
```

<h2 style="color:red">Batting Summary </h2>

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data_bats = json.load(f)

# we have multiple subsets of  battingSummary for each match. Hence lets create one single df that contains 
# data of all matches

all_data = []
for rec in data_bats:
    all_data.extend(rec['battingSummary'])
    
all_data
```

```python
df_bat = pd.DataFrame(all_data)
df_bat.tail()
```

```python
# Looking at the above data, I will start transforming the 'dismissal' column values to 'OUT' or 'NOT OUT'
# as the reason for out doesn't help our analysis

df_bat['out/not_out'] = df_bat.dismissal.apply(lambda x: 'not_out' if x == '' else 'out')
df_bat.head(15)
```

```python
df_bat.drop(['dismissal'], axis =1, inplace= True)
df_bat.head()
```

```python
# we see some non-ASCII characters in our data set so lets clean them from our df
df_bat['batsmanName'] = df_bat['batsmanName'].apply(lambda x: x.replace('†',''))

df_bat.head(20)
```

```python
# Now lets try linking the above two dataframes using the match ID
# since there is no unique value to create a connection between these 2 df's lets try connecting the match_ids
# with the team names as shown below

match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
    
match_ids_dict
```

```python
df_bat['match_id'] = df_bat['match'].map(match_ids_dict)
df_bat.head(10)
```

```python
# Now let export this final cleaned & transformed df into a CSV file

df_bat.to_csv('t20_json_files/batting_transformed_data.csv', index= False)
```

<h2 style="color:red">Bowling Summary </h2>

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data_bowl = json.load(f)

# we have multiple subsets of  battingSummary for each match. Hence lets create one single df that contains 
# data of all matches

all_bowl_data = []
for rec in data_bowl:
    all_bowl_data.extend(rec['bowlingSummary'])
    
all_bowl_data
```

```python
df_bowl = pd.DataFrame(all_bowl_data)
```

```python
df_bowl.head(20)
```

```python
df_bowl['match_id'] = df_bowl['match'].map(match_ids_dict)
df_bowl.tail(20)
```

```python
df_bowl.to_csv('t20_json_files/bowling_transformed_data.csv', index= False)
```
